```r
library(SportsAnalytics)
library(tidyverse)
library(lubridate)
library(ggthemr)
```

```r
ggthemr("dust")
```

```r
NBA1415 <- fetch_NBAPlayerStatistics("14-15")
```

```r
str(NBA1415)
```

```r
# 利用order()函數,選一種類型的資料出來,由大到小排序(decreasing)
NBAorder <- NBA1415[order(NBA1415$TotalMinutesPlayed, decreasing = T),]
head(NBAorder, 20)
```

```r
# 用subset()做欄位條件搜尋
subset(NBA1415, Position == "PG")[1:5,]
```

```r
# 用grepl()做字串條件搜尋
NBA1415[grepl("Steve", NBA1415$Name),]
```

```r
# 用ggplot畫出,四種不同類型的圖表(Exp:散佈圖、長條圖、折線圖...)
str(NBA1415$Position)
```

```r
ggplot(NBA1415, aes(x = ThreesAttempted, y = ThreesMade, color = Position)) +
    geom_line() +
    scale_colour_manual(values = c("#ffa628", "#ff7e70", "#ff73ba", "#c789f2", "#449eff")) +
    labs(title = "ThreesAttempted 對 ThreesMade 的關係")
# ThreesAttempted 對 ThreesMade 的關係
```

```r
ggplot(NBA1415, aes(x = Steals, y = TotalPoints)) +
    geom_point(aes(colour = Position)) +
    stat_smooth() +
    scale_colour_manual(values = c("#ffa628", "#ff7e70", "#ff73ba", "#c789f2", "#449eff")) +
    labs(title = "Steals 跟 總得分 的關係")
# Steals 跟 總得分 的關係
```

```r
ggplot(NBA1415, aes(x = Team, fill = Position)) + geom_bar() +
#     scale_colour_manual(values = c("#ffa628", "#ff7e70", "#ff73ba", "#c789f2", "#449eff")) +
    labs(title = "Steals 跟 總得分 的關係")
# 各隊伍人數及守備位置比例
```

```r
ggplot(NBA1415, aes(x = Position, y = Assists)) + geom_boxplot() +
    labs(title = "守備位置 對 助攻次數 的關係及平均")
```
