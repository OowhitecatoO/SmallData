```r
new_words <- c("傳染性肺炎", "醫療體系", "防疫措施")
writeLines(new_words, "new_words.txt")

stop_words <- c("在", "之", "更", "個", "未", "也", "與", "中", "的")
writeLines(stop_words, "stop_words.txt")
```

```r
cutter <- worker(
    user = "new_words.txt", stop_word = "stop_words.txt",
    bylines = FALSE
)
seg_words <- cutter[content]
seg_words
```

```r
tex_freq <- freq(seg_words)
```

```r
tex_freq <- arrange(tex_freq, desc(freq))
```

```r
head(tex_freq)
```

```r
par(family=("Microsoft YaHei"))
wordcloud(
    tex_freq$char, tex_freq$freq, min.freq = 2, 
    random.order = F, ordered.colors = F,
    colors = rainbow(nrow(tex_freq))
)
```

```r
library(wordcloud2)
wordcloud2(data = tex_freq)
```
